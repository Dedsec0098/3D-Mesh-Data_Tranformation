## Task 2: Normalize and Quantize the Mesh

In this task, we'll implement two normalization methods:
1. **Min-Max Normalization**: Scales coordinates to [0, 1] range
2. **Unit Sphere Normalization**: Centers the mesh and scales to fit within a unit sphere

Then we'll quantize the normalized meshes with a bin size of 1024.

### Step 1: Implement Normalization Methods

We'll create functions for both normalization methods.

```python
def normalize_min_max(vertices):
    """
    Min-Max Normalization: Scales all coordinates to [0, 1] range
    Formula: (x - min) / (max - min)
    
    Args:
        vertices: numpy array of shape (N, 3) containing vertex coordinates
    
    Returns:
        normalized_vertices: numpy array with coordinates in [0, 1] range
        normalization_params: dictionary containing min and max values for denormalization
    """
    vertices = np.array(vertices)
    
    # Calculate min and max for each axis
    min_vals = np.min(vertices, axis=0)
    max_vals = np.max(vertices, axis=0)
    
    # Avoid division by zero
    range_vals = max_vals - min_vals
    range_vals[range_vals == 0] = 1
    
    # Normalize to [0, 1]
    normalized_vertices = (vertices - min_vals) / range_vals
    
    # Store parameters for potential denormalization
    normalization_params = {
        'method': 'min_max',
        'min_vals': min_vals,
        'max_vals': max_vals,
        'range_vals': range_vals
    }
    
    return normalized_vertices, normalization_params


def normalize_unit_sphere(vertices):
    """
    Unit Sphere Normalization: Centers the mesh at origin and scales to fit within unit sphere
    
    Steps:
    1. Center the mesh by subtracting the centroid
    2. Scale so the farthest point is at distance 1 from origin
    
    Args:
        vertices: numpy array of shape (N, 3) containing vertex coordinates
    
    Returns:
        normalized_vertices: numpy array with mesh centered and scaled to unit sphere
        normalization_params: dictionary containing centroid and scale factor
    """
    vertices = np.array(vertices)
    
    # Calculate centroid (center of mass)
    centroid = np.mean(vertices, axis=0)
    
    # Center the vertices
    centered_vertices = vertices - centroid
    
    # Calculate maximum distance from origin
    max_distance = np.max(np.linalg.norm(centered_vertices, axis=1))
    
    # Avoid division by zero
    if max_distance == 0:
        max_distance = 1
    
    # Scale to unit sphere
    normalized_vertices = centered_vertices / max_distance
    
    # Store parameters for potential denormalization
    normalization_params = {
        'method': 'unit_sphere',
        'centroid': centroid,
        'scale_factor': max_distance
    }
    
    return normalized_vertices, normalization_params


# Test the normalization functions on one mesh
print("="*80)
print("TESTING NORMALIZATION METHODS")
print("="*80)

# Choose the first mesh for demonstration
test_mesh_name = list(meshes.keys())[0]
test_vertices = all_vertices[test_mesh_name]

print(f"\nOriginal {test_mesh_name} statistics:")
print(f"  X range: [{test_vertices[:, 0].min():.6f}, {test_vertices[:, 0].max():.6f}]")
print(f"  Y range: [{test_vertices[:, 1].min():.6f}, {test_vertices[:, 1].max():.6f}]")
print(f"  Z range: [{test_vertices[:, 2].min():.6f}, {test_vertices[:, 2].max():.6f}]")
print(f"  Centroid: {np.mean(test_vertices, axis=0)}")

# Test Min-Max Normalization
normalized_minmax, params_minmax = normalize_min_max(test_vertices)
print(f"\nAfter Min-Max Normalization:")
print(f"  X range: [{normalized_minmax[:, 0].min():.6f}, {normalized_minmax[:, 0].max():.6f}]")
print(f"  Y range: [{normalized_minmax[:, 1].min():.6f}, {normalized_minmax[:, 1].max():.6f}]")
print(f"  Z range: [{normalized_minmax[:, 2].min():.6f}, {normalized_minmax[:, 2].max():.6f}]")
print(f"  Centroid: {np.mean(normalized_minmax, axis=0)}")

# Test Unit Sphere Normalization
normalized_sphere, params_sphere = normalize_unit_sphere(test_vertices)
print(f"\nAfter Unit Sphere Normalization:")
print(f"  X range: [{normalized_sphere[:, 0].min():.6f}, {normalized_sphere[:, 0].max():.6f}]")
print(f"  Y range: [{normalized_sphere[:, 1].min():.6f}, {normalized_sphere[:, 1].max():.6f}]")
print(f"  Z range: [{normalized_sphere[:, 2].min():.6f}, {normalized_sphere[:, 2].max():.6f}]")
print(f"  Centroid: {np.mean(normalized_sphere, axis=0)}")
print(f"  Max distance from origin: {np.max(np.linalg.norm(normalized_sphere, axis=1)):.6f}")

print("\n✓ Normalization functions implemented successfully!")
```

### Step 2: Implement Quantization Function

Quantize the normalized coordinates using a bin size of 1024.

```python
def quantize_vertices(vertices, bin_size=1024):
    """
    Quantize vertex coordinates to discrete bins
    
    Args:
        vertices: numpy array of normalized vertices (assumed to be in [0, 1] or [-1, 1] range)
        bin_size: number of discrete bins (default: 1024)
    
    Returns:
        quantized_vertices: vertices with quantized coordinates
        quantization_params: dictionary containing quantization parameters
    """
    vertices = np.array(vertices)
    
    # Find the actual range of the data
    min_val = np.min(vertices)
    max_val = np.max(vertices)
    data_range = max_val - min_val
    
    # Avoid division by zero
    if data_range == 0:
        data_range = 1
    
    # Scale to [0, bin_size-1] range
    scaled = ((vertices - min_val) / data_range) * (bin_size - 1)
    
    # Quantize by rounding to nearest integer
    quantized = np.round(scaled).astype(np.int32)
    
    # Convert back to floating point in the original range
    quantized_vertices = (quantized.astype(np.float32) / (bin_size - 1)) * data_range + min_val
    
    quantization_params = {
        'bin_size': bin_size,
        'min_val': min_val,
        'max_val': max_val,
        'data_range': data_range
    }
    
    return quantized_vertices, quantization_params


# Test quantization on normalized vertices
print("="*80)
print("TESTING QUANTIZATION")
print("="*80)

print(f"\nQuantizing Min-Max normalized {test_mesh_name}:")
print(f"  Before quantization - unique X values: {len(np.unique(normalized_minmax[:, 0]))}")
quantized_minmax, quant_params_minmax = quantize_vertices(normalized_minmax, bin_size=1024)
print(f"  After quantization - unique X values: {len(np.unique(quantized_minmax[:, 0]))}")
print(f"  X range: [{quantized_minmax[:, 0].min():.6f}, {quantized_minmax[:, 0].max():.6f}]")

print(f"\nQuantizing Unit Sphere normalized {test_mesh_name}:")
print(f"  Before quantization - unique X values: {len(np.unique(normalized_sphere[:, 0]))}")
quantized_sphere, quant_params_sphere = quantize_vertices(normalized_sphere, bin_size=1024)
print(f"  After quantization - unique X values: {len(np.unique(quantized_sphere[:, 0]))}")
print(f"  X range: [{quantized_sphere[:, 0].min():.6f}, {quantized_sphere[:, 0].max():.6f}]")

print("\n✓ Quantization function implemented successfully!")
```

### Step 3: Apply Normalization and Quantization to All Meshes

Process all meshes with both normalization methods and quantize them.

```python
# Process all meshes with both normalization methods
print("="*80)
print("PROCESSING ALL MESHES")
print("="*80)

# Storage for processed meshes
normalized_meshes_minmax = {}
normalized_meshes_sphere = {}
quantized_meshes_minmax = {}
quantized_meshes_sphere = {}

# Storage for parameters
norm_params_minmax = {}
norm_params_sphere = {}
quant_params_minmax = {}
quant_params_sphere = {}

for mesh_name, vertices in all_vertices.items():
    print(f"\nProcessing {mesh_name}...")
    
    # Method 1: Min-Max Normalization
    norm_verts_minmax, params_mm = normalize_min_max(vertices)
    normalized_meshes_minmax[mesh_name] = norm_verts_minmax
    norm_params_minmax[mesh_name] = params_mm
    
    # Quantize Min-Max normalized vertices
    quant_verts_minmax, params_qmm = quantize_vertices(norm_verts_minmax, bin_size=1024)
    quantized_meshes_minmax[mesh_name] = quant_verts_minmax
    quant_params_minmax[mesh_name] = params_qmm
    
    # Method 2: Unit Sphere Normalization
    norm_verts_sphere, params_us = normalize_unit_sphere(vertices)
    normalized_meshes_sphere[mesh_name] = norm_verts_sphere
    norm_params_sphere[mesh_name] = params_us
    
    # Quantize Unit Sphere normalized vertices
    quant_verts_sphere, params_qus = quantize_vertices(norm_verts_sphere, bin_size=1024)
    quantized_meshes_sphere[mesh_name] = quant_verts_sphere
    quant_params_sphere[mesh_name] = params_qus
    
    print(f"  ✓ Min-Max: normalized and quantized")
    print(f"  ✓ Unit Sphere: normalized and quantized")

print("\n" + "="*80)
print("PROCESSING COMPLETE!")
print("="*80)
print(f"\nTotal meshes processed: {len(meshes)}")
print(f"  - {len(normalized_meshes_minmax)} Min-Max normalized meshes")
print(f"  - {len(normalized_meshes_sphere)} Unit Sphere normalized meshes")
print(f"  - {len(quantized_meshes_minmax)} Min-Max quantized meshes")
print(f"  - {len(quantized_meshes_sphere)} Unit Sphere quantized meshes")
```

### Step 6: Detailed Comparison and Analysis

Calculate metrics to compare the two normalization methods.

```python
import pandas as pd

def calculate_distortion_metrics(original_verts, processed_verts):
    """
    Calculate metrics to measure distortion between original and processed vertices
    """
    # Calculate pairwise distances in original mesh
    from scipy.spatial.distance import pdist, squareform
    
    # Sample vertices if mesh is too large (for performance)
    if len(original_verts) > 1000:
        indices = np.random.choice(len(original_verts), 1000, replace=False)
        orig_sample = original_verts[indices]
        proc_sample = processed_verts[indices]
    else:
        orig_sample = original_verts
        proc_sample = processed_verts
    
    # Calculate distances
    orig_distances = pdist(orig_sample)
    proc_distances = pdist(proc_sample)
    
    # Normalize distances to compare shape preservation
    orig_distances_norm = orig_distances / (np.max(orig_distances) + 1e-10)
    proc_distances_norm = proc_distances / (np.max(proc_distances) + 1e-10)
    
    # Calculate metrics
    distance_error = np.mean(np.abs(orig_distances_norm - proc_distances_norm))
    max_distance_error = np.max(np.abs(orig_distances_norm - proc_distances_norm))
    
    # Calculate aspect ratio preservation
    orig_ranges = np.max(original_verts, axis=0) - np.min(original_verts, axis=0)
    proc_ranges = np.max(processed_verts, axis=0) - np.min(processed_verts, axis=0)
    
    orig_aspect = orig_ranges / (np.max(orig_ranges) + 1e-10)
    proc_aspect = proc_ranges / (np.max(proc_ranges) + 1e-10)
    
    aspect_error = np.mean(np.abs(orig_aspect - proc_aspect))
    
    return {
        'mean_distance_error': distance_error,
        'max_distance_error': max_distance_error,
        'aspect_ratio_error': aspect_error
    }


print("="*80)
print("COMPARING NORMALIZATION METHODS")
print("="*80)

comparison_results = []

for mesh_name in meshes.keys():
    base_name = mesh_name.replace('.obj', '')
    original_verts = all_vertices[mesh_name]
    
    # Calculate metrics for Min-Max method
    metrics_mm_norm = calculate_distortion_metrics(original_verts, normalized_meshes_minmax[mesh_name])
    metrics_mm_quant = calculate_distortion_metrics(original_verts, quantized_meshes_minmax[mesh_name])
    
    # Calculate metrics for Unit Sphere method
    metrics_us_norm = calculate_distortion_metrics(original_verts, normalized_meshes_sphere[mesh_name])
    metrics_us_quant = calculate_distortion_metrics(original_verts, quantized_meshes_sphere[mesh_name])
    
    comparison_results.append({
        'Mesh': base_name,
        'MM_Norm_DistError': f"{metrics_mm_norm['mean_distance_error']:.6f}",
        'MM_Quant_DistError': f"{metrics_mm_quant['mean_distance_error']:.6f}",
        'US_Norm_DistError': f"{metrics_us_norm['mean_distance_error']:.6f}",
        'US_Quant_DistError': f"{metrics_us_quant['mean_distance_error']:.6f}",
        'MM_AspectError': f"{metrics_mm_norm['aspect_ratio_error']:.6f}",
        'US_AspectError': f"{metrics_us_norm['aspect_ratio_error']:.6f}"
    })

df_comparison = pd.DataFrame(comparison_results)
print("\n" + "="*100)
print(" DISTORTION METRICS COMPARISON")
print("="*100)
print("\nLegend:")
print("  MM = Min-Max Normalization")
print("  US = Unit Sphere Normalization")
print("  DistError = Mean Distance Error (lower is better)")
print("  AspectError = Aspect Ratio Preservation Error (lower is better)")
print("\n")
print(df_comparison.to_string(index=False))
print("="*100)

# Calculate average metrics
avg_mm_norm = df_comparison['MM_Norm_DistError'].astype(float).mean()
avg_us_norm = df_comparison['US_Norm_DistError'].astype(float).mean()
avg_mm_quant = df_comparison['MM_Quant_DistError'].astype(float).mean()
avg_us_quant = df_comparison['US_Quant_DistError'].astype(float).mean()

print(f"\nAVERAGE DISTANCE ERRORS:")
print(f"  Min-Max Normalized:     {avg_mm_norm:.6f}")
print(f"  Min-Max Quantized:      {avg_mm_quant:.6f}")
print(f"  Unit Sphere Normalized: {avg_us_norm:.6f}")
print(f"  Unit Sphere Quantized:  {avg_us_quant:.6f}")

print("\n✓ Comparison analysis complete!")
```

### Step 8: Written Comparison - Which Method Preserves Structure Better?

**Comparison Analysis: Min-Max vs Unit Sphere Normalization**

#### 1. **Min-Max Normalization**

**How it works:**
- Scales each coordinate independently to the [0, 1] range
- Formula: `(x - min) / (max - min)`

**Advantages:**
- ✅ **Preserves aspect ratios**: Since each axis is scaled independently based on its own range, the relative proportions are maintained
- ✅ **Predictable bounds**: All coordinates are guaranteed to be in [0, 1]
- ✅ **Simple and intuitive**: Easy to understand and implement
- ✅ **Good for rectangular/box-shaped objects**: Works well when the mesh naturally fits in an axis-aligned bounding box

**Disadvantages:**
- ❌ **Doesn't center the mesh**: The mesh may not be centered at the origin
- ❌ **Sensitive to outliers**: A single far-away vertex can compress the entire mesh
- ❌ **Axis-dependent**: Treats each axis independently, which may not be ideal for all shapes

**Best for:** Meshes with well-defined bounding boxes and uniform distribution along each axis.

---

#### 2. **Unit Sphere Normalization**

**How it works:**
- Centers the mesh at the origin by subtracting the centroid
- Scales uniformly so the farthest point is at distance 1 from the origin

**Advantages:**
- ✅ **Centers the mesh**: Always centered at origin (0, 0, 0)
- ✅ **Uniform scaling**: Scales all dimensions equally, preserving shape exactly
- ✅ **Rotation-invariant**: Results don't depend on mesh orientation
- ✅ **Natural for spherical/organic shapes**: Works well with meshes that don't align with axes
- ✅ **Better for machine learning**: Centered data often works better with neural networks

**Disadvantages:**
- ❌ **Variable bounds**: Coordinates can range anywhere in [-1, 1], depending on shape
- ❌ **May waste quantization resolution**: For elongated objects, shorter dimensions use fewer bins

**Best for:** Organic shapes, spherical objects, and machine learning applications.

---

#### 3. **Quantization Impact (1024 bins)**

Both methods are then quantized with 1024 bins:
- **Min-Max quantized**: Since the range is [0, 1], each coordinate gets mapped to 0-1023
- **Unit Sphere quantized**: The effective range depends on the shape, but typically uses full resolution

**Quantization observations:**
- Quantization introduces slight geometric distortion (vertices snap to discrete positions)
- The distortion is generally small with 1024 bins (about 0.001 unit precision)
- Both methods show similar quantization artifacts

---

#### 4. **Conclusion: Which Method is Better?**

Based on the metrics calculated above:

**🏆 Winner: Unit Sphere Normalization**

**Reasoning:**

1. **Shape Preservation**: Unit sphere normalization uses uniform scaling, which means it preserves the exact shape of the mesh. Min-max can distort proportions if the mesh has unequal dimensions.

2. **Centering**: Unit sphere places the mesh at the origin, making it easier to work with rotations and transformations.

3. **General Purpose**: Unit sphere works well across all mesh types - from elongated objects (talwar, fence) to compact ones (cylinder, explosive).

4. **ML-Friendly**: For machine learning applications, centered and uniformly scaled data typically performs better.

**When to use Min-Max instead:**
- When you need guaranteed [0, 1] bounds for a specific application
- When working with axis-aligned architectural models
- When aspect ratio distortion is acceptable for your use case

**Recommendation for this assignment:** 
**Unit Sphere Normalization** is the better choice as it preserves mesh structure more faithfully while providing consistent scaling across all dimensions.
